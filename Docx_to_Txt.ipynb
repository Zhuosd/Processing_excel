```python
# 递归获取路径下所有文件名1     # -*- coding: utf-8 -*-  
import os  
def listdir(path, list_name):  #传入存储的list
    for file in os.listdir(path):  
        file_path = os.path.join(path, file)  
        if os.path.isdir(file_path):  
            listdir(file_path, list_name)  
        else:  
            list_name.append(file_path)
```

```python
for i in range(len(listt)):
    #每个循环中docx文档和txt文档的命名
    filenamedocx = (listt[i] ) #+ '.docx')
    filenametext = (listt[i] + '.txt')
    
    #新建和打开txt文档
    f = open(filenametext,'w',encoding='utf-8')
    #打开docx的文档并读入名为file的变量中
    try:
        file = docx.Document(filenamedocx)
    except:
        continue
    #输入docx中的段落数，以检查是否空文档
    print('段落数:'+str(len(file.paragraphs)))
    #将每个段落的内容都写进去txt里面
    for para in file.paragraphs:
        f.write(para.text)
    f.close()

print('操作已经完成！')
```
